### Lab 1.1.2

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# 
img = cv.imread('img/lab1-img1.jpg')
img2 = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# 
print('X(H, W, CH)')
print(img2.shape)

# subplot 
plt.subplot(2, 2, 1)
plt.imshow(img2[:, :, 0], cmap='gray')
plt.title('Original')

#  np.transpose
tranpose_image = np.transpose(img2, (2, 0, 1))
print(f'Tranpose Y(CH, H, W) --> {tranpose_image.shape}')
plt.subplot(2, 2, 2)
plt.title("Transpose image")
plt.imshow(tranpose_image[0, :, :], cmap='gray')

#  np.moveaxis
moveaxis_image = np.moveaxis(img2, 2, 0)
print(f'Moveaxis Y(CH, H, W) --> {moveaxis_image.shape}')
plt.subplot(2, 2, 3)
plt.title("Move Axis")
plt.imshow(moveaxis_image[0, :, :], cmap="gray")

#  np.reshape
height, width, channels = img2.shape
img_reshape = np.reshape(img2, (channels, height, width))
print(f"Reshape Image Y(CH, H, W) --> {img_reshape.shape}")
plt.subplot(2, 2, 4)
plt.title('Reshape image')
plt.imshow(img_reshape[0, :, :], cmap="gray")

#  subplot
plt.subplots_adjust(wspace=0.5, hspace=0.5)

# 
plt.show()
```

### Lab 1.1.3

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# 
img = cv.imread('img/lab1-img1.jpg')
img2 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
resize_img = cv.resize(img2, (1000, 1000))
gray_img = cv.cvtColor(resize_img, cv.COLOR_RGB2GRAY)

#  (Quantization)
q_levels = 4
quantized_img = np.zeros(gray_img.shape, dtype=np.uint8)

for h in range(gray_img.shape[0]):
    for w in range(gray_img.shape[1]):
        pixel = gray_img[h, w]
        quantized_value = int((pixel / 255) * q_levels)
        quantized_img[h, w] = int(255 / q_levels) * quantized_value

# subplot  (Original)
plt.subplot(2, 2, 1)
plt.title("Original")
plt.imshow(gray_img, cmap='gray')

# subplot  (Quantization)
plt.subplot(2, 2, 2)
plt.title("Quantization")
plt.imshow(quantized_img, cmap='gray')

#  subplot
plt.subplots_adjust(wspace=0.5, hspace=0.5)

# 
plt.show()
```

### Lab 1.1.4

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('img/lab1-img1.jpg')
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
resize_img = cv.resize(gray_img, (200, 200))

#  (2D)
plt.imshow(resize_img, cmap='gray')
plt.title('2D Image')

#  (3D)
x_coords, y_coords = np.mgrid[0:resize_img.shape[0], 0:resize_img.shape[1]]
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x_coords, y_coords, resize_img, cmap='gray')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Intensity')
plt.title('3D Surface Plot')

#  subplot
plt.subplots_adjust(wspace=0.5, hspace=0.5)

# 
plt.show()
```

```python
import cv2 as cv
import matplotlib.pyplot as plt

# 
img = cv.imread("img/lab1-img1.jpg")
img2 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
RGB_img_xr = img2[:, :, 0]
RGB_img_xg = img2[:, :, 1]
RGB_img_xb = img2[:, :, 2]

#  RGB
plt.subplot(4, 4, 1)
plt.imshow(img2)
plt.title('RGB')

plt.subplot(4, 4, 2)
plt.imshow(RGB_img_xr, cmap="gray")
plt.title('R')

plt.subplot(4, 4, 3)
plt.imshow(RGB_img_xg, cmap="gray")
plt.title('G')

plt.subplot(4, 4, 4)
plt.imshow(RGB_img_xb, cmap="gray")
plt.title('B')

#  HSV  HSV
HSV_img = cv.cvtColor(img, cv.COLOR_BGR2HSV)
HSV_img_xh = HSV_img[:, :, 0]
HSV_img_xs = HSV_img[:, :, 1]
HSV_img_xv = HSV_img[:, :, 2]

plt.subplot(4, 4, 5)
plt.imshow(HSV_img)
plt.title('HSV')

plt.subplot(4, 4, 6)
plt.imshow(HSV_img_xh, cmap="gray")
plt.title('H')

plt.subplot(4, 4, 7)
plt.imshow(HSV_img_xs, cmap="gray")
plt.title('S')

plt.subplot(4, 4, 8)
plt.imshow(HSV_img_xv, cmap="gray")
plt.title('V')

#  HLS  HLS
HLS_img = cv.cvtColor(img, cv.COLOR_BGR2HLS)
HLS_img_xh = HLS_img[:, :, 0]
HLS_img_xl = HLS_img[:, :, 1]
HLS_img_xs = HLS_img[:, :, 2]

plt.subplot(4, 4, 9)
plt.imshow(HLS_img)
plt.title('HLS')

plt.subplot(4, 4, 10)
plt.imshow(HLS_img_xh, cmap="gray")
plt.title('H')

plt.subplot(4, 4, 11)
plt.imshow(HLS_img_xl, cmap="gray")
plt.title('L')

plt.subplot(4, 4, 12)
plt.imshow(HLS_img_xs, cmap="gray")
plt.title('S')

#  YCrCb  YCrCb
YCrCb_img = cv.cvtColor(img, cv.COLOR_BGR2YCrCb)
YCrCb_img_xY = YCrCb_img[:, :, 0]
YCrCb_img_xCr = YCrCb_img[:, :, 1]
YCrCb_img_xCb = YCrCb_img[:, :, 2]

plt.subplot(4, 4, 13)
plt.imshow(YCrCb_img)
plt.title('YCrCb')

plt.subplot(4, 4, 14)
plt.imshow(YCrCb_img_xY, cmap="gray")
plt.title('Y')

plt.subplot(4, 4, 15)
plt.imshow(YCrCb_img_xCr, cmap="gray")
plt.title('Cr')

plt.subplot(4, 4, 16)
plt.imshow(YCrCb_img_xCb, cmap="gray")
plt.title('Cb')

#  subplot
plt.subplots_adjust(wspace=0.6, hspace=1.2)

# 
plt.show()
```

### Lab 1.2.2

```python
import cv2 as cv
import numpy as np

img_1 = cv.imread("img/lab1-img1.jpg")
img2 = cv.cvtColor(img_1, cv.COLOR_BGR2RGB)
resize_img1 = cv.resize(img_1, (200, 200))
resize_img2 = cv.resize(img2, (200, 200))

output_file = "output/output_file.mp4"
fourcc = cv.VideoWriter_fourcc(*'mp4v')

fps = 1
frame = 100
trans_duration = 30

out = cv.VideoWriter(output_file, fourcc, fps, (resize_img1.shape[1], resize_img1.shape[0]))

w = [1.0, 0.9, 0.8, 0.7, 0.6, 0.5, 0.4, 0.3, 0.2, 0.1, 0]

#  weight 
for w1, w2 in zip(w, w[::-1]):
    image_result = cv.addWeighted(resize_img1, w1, resize_img2, w2, 0)
    for i in range(frame):
        out.write(image_result)

#  weight 
for w2, w1 in zip(w, w[::-1]):
    image_result = cv.addWeighted(resize_img1, w1, resize_img2, w2, 0)
    for i in range(frame):
        out.write(image_result)

out.release()
```

### Lab 1.2.3

```python
import cv2 as cv
import matplotlib.pyplot as plt

img_1 = cv.imread("img/lab1-img1.jpg")
img2 = cv.cvtColor(img_1, cv.COLOR_BGR2RGB)

#  1000x1000
resize_img2 = cv.resize(img2, (1000, 1000))

# 
plt.subplot(1, 3, 1)
plt.imshow(resize_img2)
plt.title('Original')

height, width = resize_img2.shape[:2]

# 
mask = np.zeros((height, width), dtype=np.uint8)
cv.rectangle(mask, (400, 150), (900, 600), 255, -1)

# 
plt.subplot(1, 3, 2)
plt.imshow(mask, cmap='gray')
plt.title('Image Mask')

# 
masked_image = cv.bitwise_and(resize_img2, resize_img2, mask=mask)

# 
plt.subplot(1, 3, 3)
plt.imshow(masked_image)
plt.title('Bitwise_AND() result')

#  subplot
plt.subplots_adjust(wspace=0.5)

# 
plt.show()
```
